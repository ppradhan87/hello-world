```
# prompt: write code to load ppt presentation

from pptx import Presentation

prs = Presentation('/content/20241001_FA35_Harbour_Audrey_revised_SIMSUB_with_TauP_Model.pptx')
```

```
for slide_number, slide in enumerate(prs.slides):
    print(f"Slide {slide_number + 1}:") # Indented this line
    for shape in slide.shapes: # Indented this line
        if hasattr(shape, "text"): # Indented this line
            print(shape.text) # Indented this line
```
